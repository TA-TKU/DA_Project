# 統一民國日期格式

```python
withDatedf.shape
```

```python
# 先把只有3碼(只有年份的)後面加上 /01/01
for i in range(0, withDatedf.shape[0]):
    tmp = withDatedf["發照日期"].iloc[i]
    if 48<=ord(tmp[0])<=57:
        if len(tmp)==3 and '00' not in tmp:
            withDatedf["發照日期"].iloc[i]=tmp+'/01/01'
        elif len(tmp) !=9:
            if '-' in tmp:
                withDatedf["發照日期"].iloc[i]=tmp.replace("-","")[0:3]+"/"+tmp.replace("-","")[3:].zfill(2)+"/01"
            else:
                withDatedf["發照日期"].iloc[i]=tmp[0:3]+"/"+tmp[3:].zfill(2)+"/01"
```

# 把所有統一格式後的民國年份改成西元年分

```python
# 把所有的民國年份改成西元年分
for i in range(0, withDatedf.shape[0]):
    tmp = withDatedf["發照日期"].iloc[i]
    tmp = tmp.replace("-","").replace(".","").replace("*","")
    if 48<=ord(tmp[0])<=57:
        if len(tmp)==3:
            withDatedf["發照日期"].iloc[i]=str(int(tmp)+1911)+'/01/01'
        elif len(tmp)==4:
            withDatedf["發照日期"].iloc[i]=str(int(tmp)+1911)+'/0%s/01' % tmp[3]
        elif len(tmp)==5:
            if 48<=ord(tmp[3:5][0])<=57:
                withDatedf["發照日期"].iloc[i]=str(int(tmp)+1911)+'/%s/01' % tmp[3:5].replace("/","")
            else:
                #print("(0):",tmp)    084//
                withDatedf["發照日期"].iloc[i]=str(int(tmp[0:3])+1911)+'/01/01'
        else:
            if len(tmp)==9 and 1<=int(tmp[4:6])<=12:
                withDatedf["發照日期"].iloc[i]=withDatedf["發照日期"].iloc[i].replace(tmp[0:3],str(int(tmp[0:3])+1911))
            elif tmp[0:2]=='00':
                temp = withDatedf["發照日期"].iloc[i].replace("/","").replace("-","").replace(' ','')[1:]
                if 1<=int(temp[3])<=12:
                    try:
                        temp = "%s/%s/%s" % (str(int(temp[0:3])+1911), temp[3].zfill(2), temp[4:].zfill(2))
                    except:
                        print("(1)第%d筆資料日期有誤%s：" %(i,tmp))
                        print(temp)
                        print('-----------------------')
                elif 1<=int(temp[3:5])<=12:
                    try:
                        temp = "%s/%s/%s" % (str(int(temp[0:3])+1911), temp[3:5].zfill(2),temp[5:].zfill(2))
                    except:
                        print("(2)第%d筆資料日期有誤%s：" %(i,tmp))
                        print(temp)      
                        print('-----------------------')
                print("更改前為%s；更改後為:%s" % (tmp,temp))
                withDatedf["發照日期"].iloc[i]=temp    
            else:
                if 0<int(tmp.split("/")[1])<=12 and 0<int(tmp.split("/")[2])<=31:
                    temp = "%s/%s/%s" % (str(int(temp[0:3])+1911), tmp.split("/")[1].zfill(2), tmp.split("/")[2].zfill(2))
                    withDatedf["發照日期"].iloc[i]=temp
                else:
                    print("(3)第%d筆資料日期有誤%s："%(i,tmp))
    else:
        print(tmp) 
```

# 哪個區域最有可能因老舊而發生火災?

```python
area_group = withDatedf.groupby(["縣市區域"])
description = area_group['發照期間'].aggregate(['min', 'max', 'mean', 'median'])
```

```python
description['mean'].nlargest(n=6) 
```

```python
description['median'].nlargest(n=6) 
```

```python
sumary=description.loc[['新北市坪林區','新北市板橋區','新北市新莊區','新北市永和區','新北市土城區','新北市三重區','新北市中和區']]
sumary
```

```python
# 尋找 眾數集中在右邊的 左（負）偏態分布   https://kknews.cc/zh-tw/news/rr22za4.html
sumary['left_skew'] = sumary['median']- sumary['mean']
sumary
```
